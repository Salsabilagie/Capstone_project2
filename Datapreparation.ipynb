MELIHAT OUTLIER

```python
df.describe()
```

```python
#Mengecek outliers
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize = (10, 5))
sns.boxplot(df['price'])
plt.annotate('Outlier', (df['price'].describe()['max'],0.1), xytext = (df['price'].describe()['max'],0.3),
             arrowprops = dict(facecolor = 'blue'), fontsize = 13 )
IQR = df['price'].describe()['75%'] - df['price'].describe()['25%']
```

```python
def find_outliers_IQR(df):
    q1=df.quantile(0.25)
    q3=df.quantile(0.75)
    IQR=q3-q1
    outliers = df[((df<(q1-1.5*IQR)) | (df>(q3+1.5*IQR)))]
    return outliers
```

```python
df.price.describe()
```

```python
outliers = find_outliers_IQR(df['price'])
print('number of outliers: '+ str(len(outliers)))
print('max outlier value: '+ str(outliers.max()))
print('min outlier value: '+ str(outliers.min()))
outliers
```

```python
#Maksimal harga (Diketahui harga airbnb yang paling mahal yaitu 835,632 thb)

Maxharga = 835632

display(len(df[df.price > Maxharga]), len(df[df.price > Maxharga])/df.shape[0]*100)
```

```python
# drop data yang melebihi harga maksimum
df.drop(index=df[df.price > Maxharga].index, inplace=True)
```

Setelah dilakukan drop data maka dilakukan kategorisasi sehingga dapat mengenai persebaran harga airbnb 

```python
#Mencari median dan MAD
PriceMed = df.price.median()
PriceMAD = abs(df.price - df.price.median()).median()

Lower = PriceMed-PriceMAD
Upper = PriceMed+PriceMAD

print(f'Median : {PriceMed}, MAD : {PriceMAD}\n Lower :{Lower}, Upper : {Upper}')
```

```python
# Pengkategorian
def Price_Category(df):
    if 2016.0 > df['price'] > 800.0 :
        return "Standar"
    elif df['price'] <= 800.0 :
        return "low"
    else:
        return "high"
     
df['Price_Category'] = df.apply(Price_Category, axis=1)
df
```
